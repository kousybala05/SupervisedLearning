```python
sample_sqft_living=df['sqft_living']
```

```python
outliers = []
def detect_outliers_iqr(data):
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    # print(q1, q3)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    # print(lwr_bound, upr_bound)
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
    return outliers# Driver code
sample_outliers = detect_outliers_iqr(sample_sqft_living)
print("Outliers from IQR method: ", sample_outliers)
```

```python
tenth_percentile = np.percentile(sample_sqft_living, 10)
ninetieth_percentile = np.percentile(sample_sqft_living, 90)
# print(tenth_percentile, ninetieth_percentile)
b = np.where(sample_sqft_living<tenth_percentile, tenth_percentile, sample_sqft_living)
b = np.where(b>ninetieth_percentile, ninetieth_percentile, b)
# print("Sample:", sample)
print("New array:",b)
```

```python
df['sqft_living_lessoutliers']=b
```
