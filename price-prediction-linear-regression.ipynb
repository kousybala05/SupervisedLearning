```python
df1.drop(['crime_rate_index'],inplace=True,axis=1)
```

```python
df1.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LinearRegression
```

```python
X=df1[['land_size_sqm', 'house_size_sqm', 'no_of_rooms', 'no_of_bathrooms',
       'large_living_room', 'parking_space', 'front_garden', 'swimming_pool','wall_fence', 'house_age','room_size']]
Y=df1[['property_value']]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33)
```

```python
model=LinearRegression()
```

```python
fit=model.fit(X_train,Y_train)
```

```python
predictions=model.predict(X_test)
```

```python
plt.scatter(Y_test,predictions)
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```
